Questão 1 \
Utilizando o dataset 'tips' crie um gráfico de barras mostrando o faturamento total do restaurante por dia, exibindo o total gasto por homens e mulheres. \
Dica: use crosstab.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
summarry = pd.crosstab(df['day'], df['sex'], values=df['total_bill'], aggfunc=np.sum)
summarry
```

```python
plt.rcParams['figure.figsize'] = [10,8]
summarry.plot(kind='bar', stacked=False)
plt.title('Restaurant total bill per sex')
plt.ylabel('Sex')
plt.xlabel('Days')
plt.show()
```

Questão 2 \
Elabre um gráfico de pizza que mostre a divisão do faturamento entre homens e mulheres.

```python
df_h = df[df['sex'] == 'Male']
df_m = df[df['sex'] == 'Female']
sizes = [df_h['total_bill'].sum(), df_m['total_bill'].sum()]
labels = ['Homens', 'Mulheres']
sns.set_style('darkgrid')
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow='True', startangle=45)
plt.title('Total_bill divison between Male and Female')
plt.axis('equal')
plt.show()
```

Questão 3 \
Elabore um gráfico de dispersão que mostre a relação entre o número de pessoas por mesa e o valor total da conta.

```python
plt.rcParams["figure.figsize"] = [10,8]
sns.set_context("notebook", font_scale=1.5, rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16})
sns.set_style("darkgrid")
sns.regplot(x='size', y='total_bill', fit_reg=True, data=df)
plt.title('Relationship between Size and total_bill')
plt.xlabel('Size')
plt.ylabel('Total_bill')
plt.show()
```
