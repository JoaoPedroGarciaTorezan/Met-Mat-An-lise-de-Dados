1) Exiba histogramas lado a lado mostrando o índice de corrupção na Europa e na África. Use bins=15.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CPI_2014.csv', encoding='latin-1')
dfEuropa = df[df['Continent'] == 'Europe']
dfAfrica = df[df['Continent'] == 'Africa']
```

```python
plt.rcParams['figure.figsize'] = [10, 6]
sns.set_style('darkgrid')

plt.subplot(1, 2, 1)
sns.histplot(dfEuropa['CPI_2014'], bins = 15, kde=False)
sns.set_style('darkgrid')
plt.title('Indice de percepção de corrupção na Europa- Ano 2014')
plt.xlabel('Nível de corrupção')
plt.axvline(60, color='k', linestyle='--')

plt.subplot(1, 2, 2)
sns.histplot(dfAfrica['CPI_2014'], bins = 15,  kde=False)
sns.set_style('darkgrid')
plt.title('Indice de percepção de corrupção na África- Ano 2014')
plt.xlabel('Nível de corrupção')
plt.axvline(60, color='k', linestyle='--')

plt.show()
```
