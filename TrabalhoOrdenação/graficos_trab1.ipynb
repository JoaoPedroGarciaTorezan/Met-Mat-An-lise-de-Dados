```python
# Importar tabelas do arquivo Word
from docx import Document
import pandas as pd
import os

# Caminho do arquivo Word
caminho_word = r'c:\Users\jpgto\OneDrive\Documentos\UNIFEI\3° Período\Met Mat Análise de Dados\Trabalho de Ordenação.docx'

# Verificar se o arquivo existe
if os.path.exists(caminho_word):
    doc = Document(caminho_word)
    print(f"Número de tabelas encontradas: {len(doc.tables)}\n")
    
    # Extrair todas as tabelas
    tabelas_df = []
    for i, table in enumerate(doc.tables):
        dados = []
        for row in table.rows:
            dados.append([cell.text for cell in row.cells])
        
        # Criar DataFrame
        df = pd.DataFrame(dados[1:], columns=dados[0])
        tabelas_df.append(df)
        print(f"Tabela {i+1}:")
        print(df)
        print("\n")
else:
    print("Arquivo não encontrado!")
```

```python
# Criar dataframes separados para cada tabela
print("\n" + "="*60)
print("Criando DataFrames separados para cada tabela")
print("="*60)

# Atribuir cada dataframe a uma variável individual
for i, df in enumerate(tabelas_df, 1):
    var_name = f'tabela{i}'
    globals()[var_name] = df
    print(f"✓ {var_name} criado com shape {df.shape}")

print(f"\nTotal de {len(tabelas_df)} tabelas carregadas e atribuídas")
print("\nVocê pode acessar os dataframes como: tabela1, tabela2, tabela3, etc...")

# Exemplo de acesso
if len(tabelas_df) > 0:
    print(f"\nExemplo - Primeiras linhas de tabela1:")
    print(globals().get('tabela1', 'Não carregado'))
```
