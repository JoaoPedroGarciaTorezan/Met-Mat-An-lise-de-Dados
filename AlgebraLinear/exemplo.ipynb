```python
import numpy as np
from numpy.linalg import norm, inv 
```

### Vetores


```python
u = np.array([3, 2])
v = np.array([1, -1])

print(f'u + v =  {u + v}') # soma
print()
print(f'2v =  {2*v}') # mult de escalar
print()
print(f'u . v =  {u.dot(v)}') # ou pode ser u@v => produto escalar
print()
print(f'||v|| =  {norm(v)}') # módulo
print()
u = (1/norm(v)) * v
print(f'||u|| =  {norm(u):.7f}') # vetor unitário
print()
```

### Matrizes

```python
Z = np.zeros([3, 3])
print(f'Matriz Nula \n={Z}')
print()
I = np.identity(3)
print(f'Matriz Identidade \n={I}')
print()
A = np.array([[1, 4, 8],
     [2, -1, 3],
     [-2, 5, 3]])
B = np.array([[4, 7, 1], 
     [3, 2, -4],
     [6, 4, -3]])
print('A + B =') # soma
print(A+B)
print()
print('2A =') # multi pelo escalar
print(2*A)
print()
print('A * B =') # multi de mattrizes
print(A@B)
print()
```

```python
F = np.array([[6, 3, 10], 
              [4, 8, 5]])

P = np.array([[0.10, 0.15],
             [0.4, 0.3], 
             [0.1, 0.2]])

print("Frutas: ")
print(F)
print()
print('Preços: ')
print(P)
print()
print("Total: ")
print(F@P)
```

### Matriz de Hilbert

```python
H = np.array([[1/ (i + j - 1)
            for j in range(1, 6)]
                for i in range(1, 6)])
print('H: ')
print(H)
print()
print('H inv: ')
print(inv(H))
print()
print('H @ inv H: ')
print(H@inv(H))
print()
print('inv(inv(H)): ')
print(inv(inv(H)))
print()
```
