### Teste de Hipótese
- Definir hipóteses Nula e Alternativa \
H0: μ = 150 ml \
Ha: μ ≠ 150ml \
Igual sempre na nula
- Probalidade de erro (Alpha)
- Teste estátistico z ou t
    z_cal = (media_amostral - media_pop)  / (dp pop * np.sqrt(numero de pop))

```python
norm.cdf(-1.96) #z_critico -> erro alpha
norm.ppf(0.025) #erro alpha ->z_critico
norm.sf(1.96) #z_critico
norm.isf(0.025) #erro alpha
```

### Teste Z
- Usado para valores n >= 30 e desvio padrão pop é conhecido \
    z_calc = (med amos - med pop) / (dp pop/np.sqrt(n))


```python
weightstats.ztest(x1=df['Machine 1'], value=150, alternative='two-sided')
# x1 = dados; value = media pop, alternative = duas caudas
# Resultado: 1- z_cal, 2- pvalue
```

### Teste t
- Usado para valores n < 30 e desvio padrão pop é desconhecido \
    t_calc = (med amos - med pop) / (dp amostral/np.sqrt(n)) \
    df = n -1

```python
volume = pd.Series([148.5, 153.4, 150.9, 151.2])
stats.ttest_1samp(volume, 150) #teste t para amostra única
# 1° valor = t-cal; 2° valor = p-value; 3° valor = degrees of freedom (n-1)
```

### Teste de uma proporção
- Tem que ter pelo menos 10 erros e 10 acertos \
    z_cal = (p - p0) / np.sqrt(p0 * (1 - p0) / n)
    p = prob atual \
    p0 = prob antiga \
- Se tem a pergunta, a proporção mudou? :\
H0: p = p0 \
Ha: p <> p0 \
- Se a pergunta é a proporção reduziu?:\
H0: p >= p0 \ 
Ha: p < p0 \
Teste de uma cauda: alpha = 0.05
- Se a pergunta é a proporção aumentou?:\
H0: p <= p0 \
Ha: p > p0 \
Teste de uma cauda: alpha = 0.05
    



```python
stats.binomtest(14, 100, p=0.21, alternative='two-sided') # retorna o p-valor do teste de proporção
#14 e 100 -> p
# p -> p0
# alterative: two-sided -> igual; less -> cauda da esquerda; greater -> cauda da direita
```

### Teste de Variância
#### Teste Chi-square
- Usado para testar variância populacional em relação a um valor específico \
Ho: s2  <= σ2 \
Ha: s2  > σ2 \
Chi_sq_cal = (n-1) * (s²) / (σ2)
#### Teste F
- Usado para testar a igualdade de duas variâncias em relação a diferentes populações
- ANOVA -> mais de duas

```python
stats.chi2.isf(0.1, 50) # erro alpha de uma cauda, df
# Res: chi_sq_critico
```

### Teste de Duas Proporções
p_linha = ((n1 * p1) + (n2 * p2)) / (n1 + n2)\
z_cal = (p1 - p2) / (np.sqrt(p_linha * (1 - p_linha) * (1 / n1 + 1 / n2)))

```python
proportion.test_proportions_2indep(30, 200, 10, 100, method='score') # zcal = statistics
```
